```python
import pandas as pd
from pycaret.classification import *
```

```python
training_data = pd.read_csv('train_data.csv')
testing_data = pd.read_csv('test_data.csv')
```

```python
training_data = training_data.iloc[:, 1:]
```

```python
qdaModel = create_model('qda')
```

```python
lrModel = create_model('lr')
```

```python
lrModel_tuned = tune_model(lrModel, n_iter=50,choose_better = True)
```

```python
lrModel_tuned1 = tune_model(lrModel, n_iter=75)
```

```python
qdaModel_tuned = tune_model(qdaModel, n_iter=50,choose_better = True)
```

```python
pred = predict_model(qdaModel_tuned, data=testing_data)
```
